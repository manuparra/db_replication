```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('results2locations.csv', header=0 ,names=['inserts','deletes','started','ended','comments','size'])
df1 = pd.read_csv('a.csv', header=0 ,names=['inserts','deletes','started','ended','comments','size'])
```

```python
df['started'] = df['started'].map(lambda x: datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S.%f'))
df['ended'] = df['ended'].map(lambda x: datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S.%f'))

df['delay'] = df['ended']- df['started']
df['delay'] = df['delay'].map(lambda x: x.total_seconds())


df1['started'] = df1['started'].map(lambda x: datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S.%f'))
df1['ended'] = df1['ended'].map(lambda x: datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S.%f'))

df1['delay'] = df1['ended']- df1['started']
df1['delay'] = df1['delay'].map(lambda x: x.total_seconds())
```
